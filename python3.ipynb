# Task
Implement a Python script that continuously asks the user to input animal names, checks if each animal is in a predefined list of allowed animals, updates separate counters for allowed and not allowed animals, and displays the status for each input. The script should allow the user to stop inputting animals and then print the final counts for both categories.

## Inicializar Contadores

### Subtask:
Inicialize as variáveis `animais_permitidos_count` e `animais_nao_permitidos_count` para 0.


**Reasoning**:
The subtask requires initializing two variables, `animais_permitidos_count` and `animais_nao_permitidos_count`, to zero. A Python code block is the appropriate way to do this.



```python
animais_permitidos_count = 0
animais_nao_permitidos_count = 0

print(f"animais_permitidos_count inicializado como: {animais_permitidos_count}")
print(f"animais_nao_permitidos_count inicializado como: {animais_nao_permitidos_count}")
```

## Definir Animais Permitidos

### Subtask:
Crie uma lista de strings contendo os nomes dos animais que serão considerados 'permitidos'.


**Reasoning**:
To define the allowed animals as requested by the subtask, I will create a Python list named `animais_permitidos` containing several animal names.



```python
animais_permitidos = ['cachorro', 'gato', 'passaro', 'peixe', 'leão', 'tigre', 'elefante']
print(f"Lista de animais permitidos criada: {animais_permitidos}")
```

## Loop de Entrada de Animais

### Subtask:
Crie um loop que continuamente pede ao usuário nomes de animais, verifica se eles são permitidos, atualiza contadores e permite que o usuário encerre o processo.


**Reasoning**:
The subtask requires implementing a loop to continuously ask for animal names, check if they are allowed, update counters, and allow the user to stop the process. This involves using a `while True` loop, `input()`, string manipulation (`.lower()`), conditional checks (`if`/`else`), and incrementing the previously initialized counters.



```python
print("Iniciando o processo de entrada de animais. Digite 'sair' para encerrar.")
while True:
    animal_digitado = input('Digite o nome de um animal (ou \'sair\' para parar): ').lower()

    if animal_digitado == 'sair':
        print("Processo encerrado.")
        break

    if animal_digitado in animais_permitidos:
        print(f"{animal_digitado} é um animal permitido.")
        animais_permitidos_count += 1
    else:
        print(f"{animal_digitado} NÃO é um animal permitido.")
        animais_nao_permitidos_count += 1
```
