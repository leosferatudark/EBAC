```python
valor_entrada = float(input('Preço de entrada no contrato: '))
risco_percentual = 0.5 # Exemplo: 0.5% de stop

# Calculando o preço de saída (Stop Loss)
valor_stop = valor_entrada - (valor_entrada * risco_percentual / 100)

print(f'Se o preço cair para {valor_stop}, caia fora da operação!')
```

isso a baixo chega a ser ridiculo, mas é incrivel!!!

```python
Se você quer um desconto de...,Multiplique por:,Explicação (O que sobra)
5%,* 0.95,Você fica com 95% do valor.
10%,* 0.90,Você fica com 90% do valor.
20%,* 0.80,Você fica com 80% do valor.
50%,* 0.50,Você fica com metade do valor.
```
