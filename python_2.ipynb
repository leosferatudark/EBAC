## F-strings: Explicação Detalhada

**F-strings** (ou *formatted string literals*) são uma das maneiras mais modernas e convenientes de formatar strings em Python, introduzidas a partir da versão 3.6.

### Como funcionam:

1.  **Prefixe com `f` ou `F`**: Para criar uma f-string, basta prefixar a string com a letra `f` (minúscula) ou `F` (maiúscula) antes das aspas (simples ou duplas).
    ```python
    nome = "Maria"
    idade = 30
    mensagem = f"Olá, meu nome é {nome} e eu tenho {idade} anos."
    print(mensagem)
    # Saída: Olá, meu nome é Maria e eu tenho 30 anos.
    ```

2.  **Use chaves `{}` para expressões**: Dentro da string, você pode incorporar expressões Python diretamente entre chaves `{}`. Qualquer coisa que você colocar dentro dessas chaves será avaliada como código Python e o resultado será inserido na string.

    *   **Variáveis**: Como visto acima (`{nome}`, `{idade}`).
    *   **Cálculos**: Você pode realizar operações matemáticas ou outras expressões.
        ```python
        preco = 10
        quantidade = 3
        total = f"O total da compra é R$ {preco * quantidade:.2f}."
        print(total)
        # Saída: O total da compra é R$ 30.00.
        ```
        (O `.2f` é um especificador de formato para duas casas decimais, mas isso é um tópico mais avançado de formatação).
    *   **Chamadas de função**: Você pode chamar funções.
        ```python
        def saudacao(nome):
            return f"Olá, {nome}"
        
        usuario = "Carlos"
        boas_vindas = f"Mensagem: {saudacao(usuario)}!"
        print(boas_vindas)
        # Saída: Mensagem: Olá, Carlos!
        ```

### Vantagens:

*   **Legibilidade**: O código é mais fácil de ler e entender, pois a expressão está diretamente no local onde seu resultado aparecerá.
*   **Concisão**: Requer menos código do que métodos de formatação mais antigos (como `%` ou `.format()`).
*   **Desempenho**: F-strings são geralmente mais rápidas do que outros métodos de formatação de strings.

Em resumo, as f-strings são uma ferramenta poderosa para criar strings dinâmicas em Python, permitindo que você misture texto fixo com resultados de expressões Python de forma clara e eficiente, usando as chaves `{}` como marcadores para o código a ser avaliado.
